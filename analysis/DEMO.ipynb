1. Самые популярные (просматриваемые) dataset'ы на данный момент?

```python
arr=[]
for i in range(len(kd1.index)):
    arr.append(kd1.iloc[i,12])
for j in range(len(arr)):
    arr[j]=arr[j].replace('k','000')
    arr[j]=int(arr[j])

kd1['Views']=arr
display(kd1.sort_values(
    'Views',
    ascending = False))
```
