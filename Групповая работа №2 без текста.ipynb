#### Некоторая подготовка таблицы к работе с ней.

Сначала добавим более удобный вариант для работы с темами/тэгами таблиц данных, преобразуя их в векторы.

````python
tags = sorted(list(set((','.join(list(df['Tags']))).split(','))))
vtags = []
for i in df.index:
    vtags.append(np.array([int(tag in (df.at[i,'Tags']).split(',')) for tag in tags]))
df['Vector Tags']=vtags
````

````python
df.at[0,'Vector Tags']
````

Аналогично сделаем с колонкой последнего обновления.

К сожалению, у сайта ```kaggle```, а как следствие у этих данных, есть одна неприятная особенность. Время указывается в не слабо пригодном формате (если точнее, то при достижении промежутка хотя бы в год, сайт округляет значения до лет).

````python
update = []
for i in df.index:
    if 'hour' in ((df.at[i,'Updated on']).split())[1]:
        time_units = [((df.at[i,'Updated on']).split())[0]]+['Hour']
    else:
        time_units = ((df.at[i,'Updated on']).split())[:2]
    if time_units[0]=='a' or time_units[0]=='an':
        time_units[0]=1
    time_units[0] = int(time_units[0])
    time_units = time_units[::-1]
    update.append(time_units)
df['Updated on'] = update
````

Переводим значения размера файла в одну единицу измерения (в байты) и делаем их тип вещественным.

````python
file_size=[]
for i in df.index:
    if (df.at[i,'FileSize'])[-2] == ' ':
        file_size.append(float((df.at[i,'FileSize'])[:-3]))
    elif (df.at[i,'FileSize'])[-2] == 'K':
        file_size.append(float((df.at[i,'FileSize'])[:-3])*1024)
    elif (df.at[i,'FileSize'])[-2] == 'M':
        file_size.append(float((df.at[i,'FileSize'])[:-3])*1024*1024)
    elif (df.at[i,'FileSize'])[-2] == 'G':
        file_size.append(float((df.at[i,'FileSize'])[:-3])*1024*1024*1024)
    else:
        file_size.append(float((df.at[i,'FileSize'])[:-3])*1024*1024*1024*1024)
df['FileSize']=file_size
````

В столбцах просмотры и ядра(?) убираем очень удобную литеру ```k```. К несчастью, из-за этой буквы точность серьёзно падает, и позже это будет заметно на графиках.

````python
view=[]
for i in df.index:
    view.append(int((df.at[i,'Views']).replace('k','000')))
df['Views']=view
````

````python
kernel=[]
for i in df.index:
    kernel.append(int((df.at[i,'Kernels']).replace('k','000')))
df['Kernels']=kernel
````

Как видно по ячейке ниже, "векторизация" данных о темах/тэгах сильно упрощает взаимодествие с ними. 
Ведь нам достаточно просто сделать следующее
$$\text{total_views_per_tags}=\sum_{i\in \text{index}}\text{views}(i)\cdot \text{V}\_\text{tag}(i),$$
если мы хотим получить абсолютное значение кол-ва просмотров, и
$$\text{total_views_per_tags_per_capita}=\cfrac{\sum\limits_{i\in \text{index}}\text{views}(i)\cdot \text{V}\_\text{tag}(i)}{\sum\limits_{i\in \text{index}}\text{V}\_\text{tag}(i)},$$
если мы хотим получить среднее значение кол-ва просмотров;<br>
где $\text{views}(i)$ &ndash; кол-во просмотров $i$-ой таблицы, а $\text{V}\_\text{tag}(i)$ &ndash; темы/тэги $i$-ой таблицы в векторном виде.<br>
Сразу невооружёным глазом видно, что были выкинуты темы, упомянаемые меньше 10-ти раз. Это было сделано, во-первых, из-за слабой репрезентативности таких данных, а, во-вторых, из-за того, что 355 столбцов данных очень неудобно читать.

````python
total_views_per_tags=sorted(list(filter(lambda y: y[1] in tuple(zip(*list(filter(lambda x: x[0]>9,list(zip(np.sum(np.array(df['Vector Tags'])),tags))))))[1],list(zip(np.sum(np.array(df['Vector Tags'])*np.array(df['Views'])),tags)))),reverse=True)
total_views_per_tags_per_capita=sorted(list(filter(lambda y: y[1] in tuple(zip(*list(filter(lambda x: x[0]>9,list(zip(np.sum(np.array(df['Vector Tags'])),tags))))))[1],list(zip(np.sum(np.array(df['Vector Tags'])*np.array(df['Views']))/np.sum(np.array(df['Vector Tags'])),tags)))),reverse=True)
````

````python
plt.barh(-np.arange(0,len(total_views_per_tags)),list(zip(*total_views_per_tags))[0],label=list(zip(*total_views_per_tags))[1])
plt.yticks(-np.arange(0,len(total_views_per_tags)),list(zip(*total_views_per_tags))[1])
plt.margins(0.0)
plt.subplots_adjust(left=0.2)
plt.xlabel('просмотры')
plt.title('Количество просмотров таблицы в зависимости от тем.')
plt.show()
````

````python
plt.barh(-np.arange(0,len(total_views_per_tags_per_capita)),list(zip(*total_views_per_tags_per_capita))[0],label=list(zip(*total_views_per_tags_per_capita))[1])
plt.yticks(-np.arange(0,len(total_views_per_tags_per_capita)),list(zip(*total_views_per_tags_per_capita))[1])
plt.margins(0.0)
plt.subplots_adjust(left=0.2)
plt.xlabel('просмотры/таблица')
plt.title('Среднее количество просмотров таблицы в зависимости от тем.')
plt.show()
````

## Кто-нибудь придумайте микровывод №2!!!

В следующих графиках будет использоваться логарифмическая шкала, ввиду особого распределения данных. Из-за этого пришлось слегка всё усложнять.

````python
label = np.logspace(np.log10(min(df['FileSize'])-1),np.log10(max(df['FileSize'])+1),75)
count = [0]*len(label)
sizes = sorted(df['FileSize'])
i=j=0
while i<len(df['FileSize']):
    if sizes[i]<label[j]:
        count[j]+=1
        i+=1
    else:
        j+=1
label_str = [np.sqrt(label[i]*label[i+1]) for i in range(len(label)-1)]+[label[-1]]
label_str = list(map(lambda y: str(round(y,0))+' B',list(filter(lambda x:x<1024,label_str)))) + list(map(lambda y: str(round(y/1024,2))+' kB',list(filter(lambda x:1024<x<(1024*1024),label_str)))) + list(map(lambda y: str(round(y/1024/1024,2))+' MB',list(filter(lambda x:(1024*1024)<x<(1024*1024*1024),label_str))))+list(map(lambda y: str(round(y/1024/1024/1024,2))+' GB',list(filter(lambda x:x>(1024*1024*1024),label_str))))
#label_str
````

````python
plt.plot(np.linspace(0,74,75),count)
plt.xticks(np.linspace(0,74,75)[::2],label_str[::2],rotation='vertical')
plt.subplots_adjust(bottom=0.2)
plt.title('Распределение размеров файлов.')
plt.axvline(x=np.sum(np.array(label)<np.mean(df['FileSize']))-1+(np.log10(np.mean(df['FileSize']))-np.log10(label[np.sum(np.array(label)<np.mean(df['FileSize']))-1]))/(np.log10(label[1])-np.log10(label[0])),color='r',label='Среднее значение')
plt.axvline(x=np.sum(np.array(label)<np.mean(np.sort(df['FileSize'])[len(df['FileSize'])//2-1:len(df['FileSize'])//2]))-1+(np.log10(np.mean(np.sort(df['FileSize'])[len(df['FileSize'])//2-1:len(df['FileSize'])//2]))-np.log10(label[np.sum(np.array(label)<np.mean(np.sort(df['FileSize'])[len(df['FileSize'])//2-1:len(df['FileSize'])//2]))-1]))/(np.log10(label[1])-np.log10(label[0])),color='g',label='Медиана')
plt.legend()
plt.text(55.5,60,str(np.round(np.mean(df['FileSize'])/1024/1024/1024,0))+' GB',rotation='vertical')
plt.text(29.5,30,str(np.round(np.mean(np.sort(df['FileSize'])[len(df['FileSize'])//2-1:len(df['FileSize'])//2])/1024/1024,2))+' MB',rotation='vertical')
plt.show()
plt.ylabel('Кол-во таблиц')
plt.xlabel('Размер файла')
````
